## Identify face in photo with pre-trained HOG face detector

```python
import sys
import dlib
from skimage import io
from PIL import Image
import matplotlib.pyplot as plt

file_name = 'test5.jpg'

image = io.imread(file_name)

face_detector = dlib.get_frontal_face_detector()

detected_faces = face_detector(image)

print(f"Found {len(detected_faces)} faces in the file {file_name}'")

win = dlib.image_window()

win.set_image(image)

for i, face_rect in enumerate(detected_faces):

    win.add_overlay(face_rect)
    

    x=face_rect.left()
    y=face_rect.top()
    x2=face_rect.right()
    y2=face_rect.bottom()  
    print(f' Face #{i+1} found at Left: {x} Top: {y} Right: {x2} Bottom: {y2}')

       
    face = Image.fromarray(image).crop((x,y,x2,y2))
    
    plt.subplot(1, len(detected_faces), i+1)
    plt.axis('off')
    plt.imshow(face)       

dlib.hit_enter_to_continue()
```

## Identify face in photo with pre-trained deep learning

```python
import sys
import dlib
from skimage import io
from PIL import Image
import matplotlib.pyplot as plt
import face_recognition_models

file_name = 'test5.jpg'

image = io.imread(file_name)

cnn_face_detection_model = face_recognition_models.cnn_face_detector_model_location()
cnn_face_detector = dlib.cnn_face_detection_model_v1(cnn_face_detection_model)

detected_faces = cnn_face_detector(image) 

print(f"Found {len(detected_faces)} faces in the file {file_name}'")

win = dlib.image_window()

win.set_image(image)

for i, face_rect in enumerate(detected_faces):

    win.add_overlay(face_rect.rect)
    

    x=face_rect.rect.left()
    y=face_rect.rect.top()
    x2=face_rect.rect.right()
    y2=face_rect.rect.bottom()  
    print(f' Face #{i+1} found at Left: {x} Top: {y} Right: {x2} Bottom: {y2}')
    
       
    face = Image.fromarray(image).crop((x,y,x2,y2))
    
    plt.subplot(1, len(detected_faces), i+1)
    plt.axis('off')
    plt.imshow(face)       

dlib.hit_enter_to_continue()
```
